```python
from ipynb.fs.defs.Preprocessing_Module import read_data, select_columns, reshape_dataset, fillna_dataset, calculate_diff, cdf_plot,add_topological
from ipynb.fs.defs.Feature_Expansion_Module import feature_expansion
from ipynb.fs.defs.Initializer_Module import initializer
from ipynb.fs.defs.Hyperparameters_Module import parameters
from ipynb.fs.defs.Feature_Selection_Module import select_features
from ipynb.fs.defs.Model_Validation_Module import test_model_random_split, train_model_KFold, operational_validation, validation_with_flag
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
np.set_printoptions(suppress=True)
```

```python
data = read_data('../Datasets/Italy/Veneto/Culex/IT_veneto_culex_2010_2022.csv')
#=====================================================================
data = data[data.trap_type != 'Gravid Trap']
data = data.reset_index(drop=True)
#=====================================================================
columns = ['station_id', 'dt_placement', 'nuts3_name',
           'culex.spp', 'ndvi', 'ndwi', 'ndmi', 'ndbi',
           'ndvi_mean', 'ndvi_std', 'ndwi_mean', 'ndwi_std', 'ndmi_mean',
           'ndmi_std', 'ndbi_mean', 'ndbi_std', 'lst_day', 'lst_night',
           'acc_rainfall_1week', 'acc_rainfall_2week', 'acc_rainfall_jan',
           'lst_jan_mean', 'lst_feb_mean',
           'lst_mar_mean', 'lst_apr_mean', 'wc_l_1km',
           'wc_mean_dist_1000', 'coast_mean_dist_1000', 'dem_mean_1km',
           'slope_mean_1km', 'aspect_mean_1km', 'waw_mean_1km', 'flow_accu_1km', 'x', 'y','landcover']
names = ['station_id', 'dt_placement', 'province',
         'mosq_now', 'ndvi', 'ndwi', 'ndmi', 'ndbi',
         'ndvi_mean', 'ndvi_std', 'ndwi_mean', 'ndwi_std', 'ndmi_mean',
         'ndmi_std', 'ndbi_mean', 'ndbi_std', 'lst_day', 'lst_night',
         'acc_rainfall_1week', 'acc_rainfall_2week', 'acc_rainfall_jan',
         'lst_jan_mean', 'lst_feb_mean',
         'lst_mar_mean', 'lst_apr_mean', 'wc_l_1km',
         'wc_mean_dist_1000', 'coast_mean_dist_1000', 'dem_mean_1km',
         'slope_mean_1km', 'aspect_mean_1km', 'waw_mean_1km', 'flow_accu_1km',  'x', 'y','landcover']
data = add_topological(data,'../Datasets/Topological Datasets/Culex_Italy.csv',landcover=True,minimize=True)
data = select_columns(data,columns,names)
data = reshape_dataset(data,['x','y','dt_placement'])
data = fillna_dataset(data,{'acc_rainfall_1week':0, 'acc_rainfall_2week':0, 'acc_rainfall_jan':0})
time = calculate_diff(data,id_column='station_id',date_column='dt_placement')
step = cdf_plot(time[:6000], len(time))
```

```python
# set the number of days for the prediction based on the plot above
# step= 15
# set the 'only_env' as True if you want to use ONLY environmental features
only_env = True
# set the case title
case_title = 'Mosquito Regression'
# set the 'exp' as True if you want to save the results as csv
exp = False
# set the filepath to export the data
path = '../Resuts/Italy/Veneto/Culex/'
# set the name of the station ID column
station_col = 'station_id'
# set the name of the date column
date_col = 'dt_placement'
# set the target_type
target_type = 'mosquito'
# set the transformation
transform = False
```

```python
data = feature_expansion(data, one_hot=['province'], imput_list=[date_col,station_col], env=only_env)
```

```python
data = initializer(data, 10, 25, step, del_list=[station_col], env=only_env, date_column = date_col, mosq_column ='mosq_now')
```

```python
depth, est = parameters(data, num=10, flag=True, date_col=date_col, target_type=target_type, transform=transform)
```

```python
selected_features, depth, est = select_features(data, depth, est, grid=False, date_col=date_col,
                                                target_type=target_type, transform=transform)
```

```python
print('max_depth: ', depth)
print('Number of estimators: ', est)
print('Selected_features: ', selected_features)
```

```python
test_model_random_split(data, selected_features, depth, est, step, case=case_title, filepath = path,
                        date_col = date_col, target_type=target_type, transform=transform, export= exp)
```

```python
train_model_KFold(data,selected_features, depth, est, target_type=target_type, transform=transform, date_col=date_col)
```

```python
operational_validation(data, selected_features, depth, est, step, filepath = path,
                       date_col = date_col, date = '2022-01-01', case = case_title,
                       target_type=target_type, transform=transform, export= exp)
```
