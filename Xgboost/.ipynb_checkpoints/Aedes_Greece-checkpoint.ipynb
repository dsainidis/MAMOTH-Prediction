```python
data = read_data('/home/iro/Desktop/noa/Culex Greece/aedes.csv')
columns = ['stationid', 'dt_placeme', 'Aedes', 'X', 'Y', 'Rice_1000', 'Rice_3000',
       'Rice_5000', 'ndwi', 'ndmi', 'ndbi', 'water_1000m', 'lst_day',
       'lst_night', 'LST_Jan_mean', 'LST_Feb_mean', 'LST_Mar_mean',
       'LST_Apr_mean', 'max_wind', 'min_wind', 'mean_wind',
       'acc_rainfall_1week', 'acc_rainfall_2week2', 'acc_rainfall_jan',
       'SP_CNT', 'BIOL_CNT', 'WC_L', 'PG_AREA', 'FS_AREA', 'WC_dist',
       'Coast_dist', 'DEM', 'SLOPE', 'ASPECT', 'WAW', 'FLOW', 'MAX_TEMP',
       'MIN_TEMP', 'AVG_WIND_SPEED', 'CURR_RAIN', 'RAIN_7DAYS', 'RAIN_14DAYS',
       'RAIN_1stJAN']
names = ['stationid', 'dt_placeme', 'mosq_now', 'x', 'y', 'Rice_1000', 'Rice_3000',
       'Rice_5000', 'ndwi', 'ndmi', 'ndbi', 'water_1000m', 'lst_day',
       'lst_night', 'LST_Jan_mean', 'LST_Feb_mean', 'LST_Mar_mean',
       'LST_Apr_mean', 'max_wind', 'min_wind', 'mean_wind',
       'acc_rainfall_1week', 'acc_rainfall_2week2', 'acc_rainfall_jan',
       'SP_CNT', 'BIOL_CNT', 'WC_L', 'PG_AREA', 'FS_AREA', 'WC_dist',
       'Coast_dist', 'DEM', 'SLOPE', 'ASPECT', 'WAW', 'FLOW', 'MAX_TEMP',
       'MIN_TEMP', 'AVG_WIND_SPEED', 'CURR_RAIN', 'RAIN_7DAYS', 'RAIN_14DAYS',
       'RAIN_1stJAN']
data = select_columns(data,columns,names)
data = reshape_dataset(data,['stationid','dt_placeme','mosq_now'],[True, True, False],['stationid','dt_placeme'])
time = calculate_diff(data,id_column='stationid',date_column='dt_placeme')
```

```python
cdf_plot(time[:2600], len(time))
```

```python
#set the number of days for the prediction based on the plot above
step= 30
# set the only_env as True if you want to use only environmental features
only_env=False
```

```python
data = feature_expansion(data, imput_list=['dt_placeme','stationid'], env=only_env)
```

```python
data = initializer(data,10,20,60,step,['stationid(t)'],flag=True, env=only_env)
```

```python
depth, est = parameters(data,num=10,flag=True)
```

```python
selected_features, depth, est = select_features(data,depth,est,grid=False)
```

```python
print('max_depth: ', depth)
print('Number of estimators: ', est)
print('Selected_features: ', selected_features)
```

```python
test_model_random_split(data, selected_features, depth, est, step, case='Aedes Greece', export= False,
                        filepath = '/home/iro/Desktop/noa/Culex Greece/')
```
