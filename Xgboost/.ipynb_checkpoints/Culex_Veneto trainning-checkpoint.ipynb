## Preprocessing

```python
# format the date column as datetime object with "%Y-%m-%d" type format
data['dt_placement'] = pd.to_datetime(data['dt_placement'], format="%Y-%m-%d")

# fill the nan values of each of the two columns with the mean temerature value of the corresponding month in the region
# Hint: use fillna() and groupby() functions
data['lst_night'] = data['lst_night'].fillna(data.groupby(data['dt_placement'].dt.to_period('M'))['lst_night'].transform('mean'))
data['lst_day'] = data['lst_day'].fillna(data.groupby(data['dt_placement'].dt.to_period('M'))['lst_day'].transform('mean'))

# drop all the rows with at least one nan value
data = data.dropna().reset_index(drop=True)

data
```

## Feature extraction

```python
# calculate the average value of day(lst_day) and night(lst_night) temperature(lst)
data['lst'] = (data['lst_day'] + data['lst_night'])/2 

# extract the year feature out of dt_placement
data['year'] = data['dt_placement'].dt.year

# calculate the euclidean distance from equator
data['vert_distance'] = data.apply(lambda x: math.sqrt((x['x']-x['x'])**2 + (x['y'])**2),axis=1)

# calculate the euclidean distance from null island (0,0)
data['null_island_distance'] = data.apply(lambda x: math.sqrt((x['x'])**2 + (x['y'])**2),axis=1)

#create new features of your own
```

## Feature Grouping

```python
satellite_data = ['ndvi', 'ndwi', 'ndmi', 'ndbi', 'ndvi_mean', 'ndvi_std', 'ndwi_mean',
                  'ndwi_std', 'ndmi_mean', 'ndmi_std', 'ndbi_mean', 'ndbi_std',
                  'lst_day', 'lst_night', 'lst', 'acc_rainfall_1week', 'acc_rainfall_2week',
                  'acc_rainfall_jan', 'lst_jan_day_mean', 'lst_feb_day_mean',
                  'lst_mar_day_mean', 'lst_apr_day_mean', 'lst_jan_night_mean',
                  'lst_feb_night_mean', 'lst_mar_night_mean', 'lst_apr_night_mean']

topological_data = ['DISTANCE_TO_COAST', 'DISTANCE_TO_RIVER',
                    'SLOPE_mean_1km', 'ASPECT_mean_200m', 'ELEVATION_mean_1km',
                    'HILLSHADE_mean_1km', 'FS_AREA_1km', 'FLOW_ACCU_200m']

landcover_data = ['landcover']

misc_data = ['x', 'y', 'null_island_distance', 'vert_distance', 
             'days_distance', 'mo_sin', 'mo_cos', 'year', 'summer_days_month']
```

## Dataset Construction

```python
# Construct your own dataset with feature groups of your choice
dataset = data[satellite_data+misc_data] # e.g. 

#add target variable to the end of the dataset (do not change the below code line)
dataset = pd.concat([dataset, data['mosq_now']], axis=1)

#print the formed dataset
dataset
```

## Train and validate a prediction model

```python
# define feature set
X = dataset.iloc[:,:-1]

#define target variable
y = dataset.iloc[:,-1]

# Randomly split the above dataset in train and test set (75-25% size of the intitial dataset correspondingly)
# Hint: Use Sklearn function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)

#re-index the new sets
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

```python
# Hint: See Sklearn's documentation about DecisionTreeRegressor

# criterion = {'squared_error', 'absolute_error'}
clf = DecisionTreeRegressor(random_state=1, max_depth=10, criterion = 'squared_error', min_samples_split=100)

# Train Decision Tree Regressor
clf = clf.fit(X_train,y_train)

# Predict the response for test dataset
y_pred = clf.predict(X_test)

# Plot the trained decision tree
fig = plt.figure(figsize=(80,40))
plot_tree(clf,feature_names=X_train.columns.tolist())
plt.show()
```

```python
# Print snapshot of predictions
print(y_pred)

# Is there something to seem a little odd according to your intuition?
# Why is that?

# Round predictions to get integer values 
y_pred = np.round(y_pred)

# Print the class distribution of the predictions
y_pred = pd.Series(y_pred)
print(y_pred.value_counts().sort_index())
```

```python
# Can you think of a similar problem?
# What if there were negative or greater than 9 predictions?
y_pred[y_pred>9] = 9
y_pred[y_pred<0] = 0
```

```python
# Calculate MAE error on test set
print('MAE on test set: ',metrics.mean_absolute_error(y_test, y_pred))

# Calculate percentage of error <= 3
perc = ((np.abs(y_test-y_pred) <= 3).mean())*100
print('Error < 3:',"%.2f"%perc,'%')

# What other metric can you think of?
```

## Validation metrics and plots

```python
#define function to plot i) error per class ii) error distributiom

def plot_per_class(y_test,y_pred):
    #plot error per class
    labels = y_test.unique().tolist()
    labels.sort()
    f = []
    length = []
    y_test = y_test.reset_index(drop=True)
    for k in labels:
        cc = y_test.loc[y_test==k].index.tolist()
        length.append(len(cc))
        actual = y_test.loc[cc]
        predictions = y_pred[cc]
        mae_class = metrics.mean_absolute_error(actual, predictions)
        f.append(mae_class)
    labels = [str(int(e)) for e in labels]
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(labels,f)
    for i, v in enumerate(f):
        ax.text(i, v, str('%.2f'%(v)), rotation=30)
    plt.xlabel('class')
    plt.ylabel('MAE')
    plt.title('MAE per class \n')
    plt.show()

def error_distribution(y_test,y_pred):
    # plot error distribution
    error = np.abs(y_test-y_pred).tolist()
    bins = np.arange(len(y_test.unique())) - 0.5
    plt.hist(error, bins)
    plt.xticks(range(len(y_test.unique())))
    plt.title('Error Histogram')
    plt.xlabel('Error')
    plt.grid()
    plt.show()
    
def actual_predicted_distribution(y_test, y_pred):
    # plot error distribution
    bins = np.arange(len(y_test.unique())+1) - 0.5
    plt.hist(y_test, bins, label='Actual',alpha=0.5)
    plt.hist(y_pred, bins, label='Predicted',alpha=0.5)
    plt.xticks(range(len(y_test.unique())))
    plt.title('Predicted vs Actual Histogram')
    plt.xlabel('Error')
    plt.legend()
    plt.grid()
    plt.show()
    
def plots(y_test,y_pred):
    plot_per_class(y_test,y_pred)
    error_distribution(y_test,y_pred)
    actual_predicted_distribution(y_test,y_pred)
```

```python
plots(y_test,y_pred)
```

```python
# define a function to plot feature importance
def plot_feature_importance(feature_names, feature_importance):
    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)

    #Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)

    #Define size of bar plot
    plt.figure(figsize=(10,8))

    #Plot Searborn bar chart
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])

    #Add chart labels
    plt.xlabel('Feature Importance')
    plt.ylabel('Feature Names')
    plt.show()
```

```python
# plot feature importance of your model
feature_names = np.array(X_train.columns)
fi = clf.feature_importances_

plot_feature_importance(feature_names, fi)
```
