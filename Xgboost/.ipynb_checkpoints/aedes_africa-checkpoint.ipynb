```python
from ipynb.fs.defs.Preprocessing_Module import read_data, add_topological, select_columns, reshape_dataset, calculate_diff, cdf_plot
from ipynb.fs.defs.Feature_Expansion_Module import feature_expansion
from ipynb.fs.defs.Initializer_Module import initializer
from ipynb.fs.defs.Hyperparameters_Module import parameters
from ipynb.fs.defs.Feature_Selection_Module import select_features
from ipynb.fs.defs.Model_Validation_Module import test_model_random_split, train_model_KFold, operational_validation, validation_with_flag
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
np.set_printoptions(suppress=True)
```

```python
data = read_data('../Datasets/IT_Trentino_aedes_albopictus_entomological_data_2011_2021.csv')
columns = ['station_id', 'dt_placement', 'NUTS3_NAME',
           'Aedes albopictus', 'ndvi', 'ndwi', 'ndmi', 'ndbi',
           'ndvi_mean', 'ndvi_std', 'ndwi_mean', 'ndwi_std', 'ndmi_mean',
           'ndmi_std', 'ndbi_mean', 'ndbi_std', 'lst_day', 'lst_night','lst_jan_mean',
           'lst_feb_mean','lst_mar_mean','lst_apr_mean','acc_rainfall_1week', 'acc_rainfall_2week',
           'acc_rainfall_jan', 'x', 'y']
names = ['station_id', 'dt_placement', 'province',
         'mosq_now', 'ndvi', 'ndwi', 'ndmi', 'ndbi',
         'ndvi_mean', 'ndvi_std', 'ndwi_mean', 'ndwi_std', 'ndmi_mean',
         'ndmi_std', 'ndbi_mean', 'ndbi_std', 'lst_day', 'lst_night', 'lst_jan_mean',
         'lst_feb_mean','lst_mar_mean','lst_apr_mean','acc_rainfall_1week', 'acc_rainfall_2week',
         'acc_rainfall_jan', 'x', 'y']
# data = add_topological(data,'../Datasets/Topological Datasets/Culex_Italy.csv')
data = select_columns(data,columns,names)
data = reshape_dataset(data,['station_id','dt_placement','mosq_now'],[True, True, False],['station_id','dt_placement'])
data = data.reset_index(drop=True)
time = calculate_diff(data,id_column='station_id',date_column='dt_placement')
#=====================removal of stations with only zero mosquitoes=====================================
st = np.unique(data['station_id'])
c = []
for x in st:
    v = data.loc[data['station_id']== x]['mosq_now']
    if len(np.unique(v)) == 1:
        c.append(x)
data = data[~data.station_id.isin(c)]
data = data.reset_index(drop=True)
```

```python
step = cdf_plot(time[:1200], len(time))
```

```python
# set the number of days for the prediction based on the plot above
# step= 15
# set the only_env as True if you want to use only environmental features
only_env = False
# set the case title
case_title = 'Aedes Trentino'
# set the exp as True if you want to save the results as csv
exp = False
# set the filepath to export the data
path = '../Resuts/Culex Italy/'
# set the name of the station ID column
station_col = 'station_id'
# set the name of the date column
date_col = 'dt_placement'
```

```python
data = feature_expansion(data,one_hot=['province'], imput_list=[date_col,station_col], env=only_env)
```

```python
data = initializer(data,[-1,   1., 2.,  3., 5., 8.,  11., 14., 23., 40., np.inf],10,25,step,[station_col], env=only_env,
                   station_column = station_col, date_column = date_col)
```

```python
depth, est = parameters(data,num=10,flag=True, date_col=date_col)
```

```python
selected_features, depth, est = select_features(data,depth,est,grid=False, date_col=date_col)
```

```python
print('max_depth: ', depth)
print('Number of estimators: ', est)
print('Selected_features: ', selected_features)
```

```python
test_model_random_split(data, selected_features, depth, est, step, case=case_title, filepath = path,
                        date_col = date_col, export= exp)
```

```python
train_model_KFold(data,selected_features, depth, est, date_col=date_col)
```

```python
operational_validation(data, selected_features, depth, est, step, date='2021-01-01', filepath = path, date_col = date_col,
                       case = case_title, export= exp)
```
