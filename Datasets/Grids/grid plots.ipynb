```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
xMin, yMin, xMax, yMax = 10.26127, 44.66172, 13.78590, 46.26450
fig, ax = plt.subplots()
data_veneto = pd.read_csv('IT_veneto_shapefile_2km.csv')
plt.scatter(data_veneto['x'],data_veneto['y'])
data_trentino = pd.read_csv('IT_trentino_shapefile_2km.csv')
plt.scatter(data_trentino['x'],data_trentino['y'])
rect = patches.Rectangle((xMin, yMin), xMax-xMin, yMax-yMin,  linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)
xMin, yMin, xMax, yMax = 10.466018, 44.801067, 13.900546, 46.68757
rect = patches.Rectangle((xMin, yMin), xMax-xMin, yMax-yMin,  linewidth=1, edgecolor='black', facecolor='none')
ax.add_patch(rect)
plt.show()
data = pd.concat([data_veneto,data_trentino],axis=0).reset_index(drop=True)
data['x'] = round(data['x'],6)
data['y'] = round(data['y'],6)
print('xMin, yMin, xMax, yMax :', data['x'].min(),data['y'].min(),data['x'].max(),data['y'].max())
```
