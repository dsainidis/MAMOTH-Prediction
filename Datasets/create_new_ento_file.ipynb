```python
import pandas as pd
import os
import missingno

enc="utf-8"

pd.set_option('display.max_columns', None)
```

```python
old_file_name = 'Serbia/Vojvodina/SER_vojvodina_culex_2010_2023.csv'
```

```python
new_file_name = old_file_name.replace('2023', '2024')
new_file_name
```

```python
file_dir = os.path.dirname(old_file_name)
```

```python
data_old = pd.read_csv(old_file_name, encoding=enc, low_memory=False)
```

```python
data_old
```

```python
missingno.matrix(data_old)
```

```python
directory = file_dir + '/monthly_data/2024/'

# List all files that contain 'ento_2024' in their name
files = [f for f in os.listdir(directory) if 'ento_2024' in f]

print(files)
```

```python
file_paths = [directory + file for file in files]
file_paths
```

```python
data_new = [pd.read_csv(file, encoding=enc) for file in file_paths]
```

```python
merged_new = pd.concat(data_new, ignore_index=True, sort=False)
```

```python
merged_new
```

```python
missingno.matrix(merged_new)
```

```python
is_match = False
# 1. Check if columns match (ignoring order)
if set(data_old.columns) == set(merged_new.columns):
    print("✅ Columns match! Rearranging merged_new...")
    is_match = True
    
    # 2. Rearrange merged_new columns to match the order of data_old
    merged_new = merged_new[data_old.columns]
    
else:
    print("❌ Columns do NOT match!")
    missing_in_new = set(data_old.columns) - set(merged_new.columns)
    missing_in_old = set(merged_new.columns) - set(data_old.columns)
    print("Missing in merged_new:", missing_in_new)
    print("Missing in data_old:", missing_in_old)
```

```python
if not is_match:
    # 1. Find columns missing in merged_new (present in data_old but missing in merged_new)
    missing_in_new = set(data_old.columns) - set(merged_new.columns)

    # 2. Find extra columns in merged_new (present in merged_new but not in data_old)
    extra_in_new = set(merged_new.columns) - set(data_old.columns)

    # 3. Drop extra columns
    if extra_in_new:
        merged_new = merged_new.drop(columns=list(extra_in_new))

    # 4. Add missing columns (filled with NaN)
    for col in missing_in_new:
        merged_new[col] = pd.NA  # or use np.nan if you prefer

    # 5. Rearrange columns to match the order in data_old
    merged_new = merged_new[data_old.columns]

    print("✅ merged_new now matches data_old structure!")
```

```python
comparison = pd.DataFrame({'data_old': data_old.dtypes, 'merged_new': merged_new.dtypes})

# Find mismatches
mismatches = comparison[comparison['data_old'] != comparison['merged_new']]

print(mismatches)
```

```python
for col in mismatches.index:
    target_dtype = data_old[col].dtype
    
    try:
        # Try standard astype
        merged_new[col] = merged_new[col].astype(target_dtype)
    except TypeError:
        # If fails (e.g., because of NAType), use a safe conversion
        if pd.api.types.is_numeric_dtype(target_dtype):
            merged_new[col] = pd.to_numeric(merged_new[col], errors='coerce')
        else:
            merged_new[col] = merged_new[col].astype(str)  # fallback to string if needed
```

```python
match = (data_old.dtypes == merged_new.dtypes).all()
print("Do they match now? ", match)
```
