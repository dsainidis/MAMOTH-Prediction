```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as mtick
from matplotlib_venn import venn2,venn3
```

```python
italy = pd.read_csv('total_veneto_culex.csv')
italy['error'] = italy['actual'] - italy['predictions']
italy['abs_error'] = np.abs(italy['predictions'] - italy['actual'])
italy['dt_placement'] = pd.to_datetime(italy['dt_placement'], format="%Y-%m-%d")
italy['dt_prediction'] = pd.to_datetime(italy['dt_prediction'], format="%Y-%m-%d")
italy['region'] ='italy'
italy
```

```python
germany = pd.read_csv('total_germany_culex.csv')
germany['error'] = germany['actual']-germany['predictions']
germany['abs_error'] = np.abs(germany['predictions'] - germany['actual'])
germany['dt_placement'] = pd.to_datetime(germany['dt_placement'], format="%Y-%m-%d")
germany['dt_prediction'] = pd.to_datetime(germany['dt_prediction'], format="%Y-%m-%d")
germany['region'] = 'Germany'
germany
```

```python
greece = pd.read_csv('total_CM_culex.csv')
greece['error'] =  greece['actual'] - greece['predictions']
greece['abs_error'] = np.abs(greece['predictions'] - greece['actual'])
greece['dt_placement'] = pd.to_datetime(greece['dt_placement'], format="%Y-%m-%d")
greece['dt_prediction'] = pd.to_datetime(greece['dt_prediction'], format="%Y-%m-%d")
greece['region'] = 'Greece'
greece
```

```python
serbia = pd.read_csv('total_serbia_culex.csv')
serbia['error'] =  serbia['actual'] - serbia['predictions']
serbia['abs_error'] = np.abs(serbia['predictions'] - serbia['actual'])
serbia['dt_placement'] = pd.to_datetime(serbia['dt_placement'], format="%Y-%m-%d")
serbia['dt_prediction'] = pd.to_datetime(serbia['dt_prediction'], format="%Y-%m-%d")
serbia['region'] = 'serbia'
serbia
```

```python
overall = pd.concat([italy, greece, germany, serbia],axis = 0)
len_overall = len(overall)
len_ent = len(overall[overall['entomological_features']==True])
len_env = len(overall[overall['entomological_features']==False])
overall
```

```python
count_df = overall.groupby(overall.dt_prediction.dt.year).count()['predictions']
ent_df = overall[overall['entomological_features']==True].groupby(overall[overall['entomological_features']==True].dt_prediction.dt.year).count()['predictions']
env_df = overall[overall['entomological_features']==False].groupby(overall[overall['entomological_features']==False].dt_prediction.dt.year).count()['predictions']
count_df = pd.concat([count_df, ent_df, env_df], axis=1).reset_index()
count_df.columns = ['year','overall', 'ent', 'env']
print(count_df)

fig = plt.figure()
ax=count_df.plot(kind='bar', x='year', rot=0, ylabel="Number of predictions", xlabel="Year")
i = 0
for p in ax.patches:
    ax.annotate('%.0f'%(p.get_height()), (p.get_x() * 1.035, p.get_height() * 1.005),rotation=35)
    i += 1
ax.set_title("Number of predctions per Year")
```

```python
years = ['2020', '2021', '2022']
it_overall = []
it_env = []
it_ent = []
ger_overall = []
ger_env = []
ger_ent = []
gre_overall = []
gre_env = []
gre_ent = []
ser_overall = []
ser_env = []
ser_ent = []

it_overall_perc = []
it_env_perc = []
it_ent_perc = []
ger_overall_perc = []
ger_env_perc = []
ger_ent_perc = []
gre_overall_perc = []
gre_env_perc = []
gre_ent_perc = []
ser_overall_perc = []
ser_env_perc = []
ser_ent_perc = []

it_overall_std = []
it_env_std = []
it_ent_std = []
ger_overall_std = []
ger_env_std = []
ger_ent_std = []
gre_overall_std = []
gre_env_std = []
gre_ent_std = []
ser_overall_std = []
ser_env_std = []
ser_ent_std = []

for x in years:
    pred = italy[italy['dt_prediction'].dt.year==int(x)]
    it_overall.append(pred['abs_error'].mean())
    it_overall_perc.append((pred['abs_error']<=3).mean())
    pred_env = pred[pred['entomological_features']==False]
    it_env.append(pred_env['abs_error'].mean())
    it_env_perc.append((pred_env['abs_error']<=3).mean())
    pred_ent = pred[pred['entomological_features']==True]
    it_ent.append(pred_ent['abs_error'].mean())
    it_ent_perc.append((pred['abs_error']<=3).mean())
    
    it_overall_std.append(pred['abs_error'].std())
    pred_env = pred[pred['entomological_features']==False]
    it_env_std.append(pred_env['abs_error'].std())
    pred_ent = pred[pred['entomological_features']==True]
    it_ent_std.append(pred_ent['abs_error'].std())
    
    pred = germany[germany['dt_prediction'].dt.year==int(x)]
    ger_overall.append(pred['abs_error'].mean())
    ger_overall_perc.append((pred['abs_error']<=3).mean())
    pred_env = pred[pred['entomological_features']==False]
    ger_env.append(pred_env['abs_error'].mean())
    ger_env_perc.append((pred_env['abs_error']<=3).mean())
    pred_ent = pred[pred['entomological_features']==True]
    ger_ent.append(pred_ent['abs_error'].mean())
    ger_ent_perc.append((pred['abs_error']<=3).mean())
    
    ger_overall_std.append(pred['abs_error'].std())
    pred_env = pred[pred['entomological_features']==False]
    ger_env_std.append(pred_env['abs_error'].std())
    pred_ent = pred[pred['entomological_features']==True]
    ger_ent_std.append(pred_ent['abs_error'].std())
    
    pred = greece[greece['dt_prediction'].dt.year==int(x)]
    gre_overall_perc.append((pred['abs_error']<=3).mean())
    gre_overall.append(pred['abs_error'].mean())
    pred_env = pred[pred['entomological_features']==False]
    gre_env.append(pred_env['abs_error'].mean())
    gre_env_perc.append((pred_env['abs_error']<=3).mean())
    pred_ent = pred[pred['entomological_features']==True]
    gre_ent.append(pred_ent['abs_error'].mean())
    gre_ent_perc.append((pred['abs_error']<=3).mean())
    
    gre_overall_std.append(pred['abs_error'].std())
    pred_env = pred[pred['entomological_features']==False]
    gre_env_std.append(pred_env['abs_error'].std())
    pred_ent = pred[pred['entomological_features']==True]
    gre_ent_std.append(pred_ent['abs_error'].std())
    
    pred = serbia[serbia['dt_prediction'].dt.year==int(x)]
    ser_overall.append(pred['abs_error'].mean())
    ser_overall_perc.append((pred['abs_error']<=3).mean())
    pred_env = pred[pred['entomological_features']==False]
    ser_env.append(pred_env['abs_error'].mean())
    ser_env_perc.append((pred_env['abs_error']<=3).mean())
    pred_ent = pred[pred['entomological_features']==True]
    ser_ent.append(pred_ent['abs_error'].mean())
    ser_ent_perc.append((pred['abs_error']<=3).mean())
    
    ser_overall_std.append(pred['abs_error'].std())
    pred_env = pred[pred['entomological_features']==False]
    ser_env_std.append(pred_env['abs_error'].std())
    pred_ent = pred[pred['entomological_features']==True]
    ser_ent_std.append(pred_ent['abs_error'].std())
```

```python
df = pd.DataFrame()
df['year'] =years
df['overall'] = [np.nanmean([a,b,c,d]) for a, b, c, d in zip(it_overall, ger_overall, gre_overall, ser_overall)]
df['ent'] = [np.nanmean([a,b,c,d]) for a, b, c, d in zip(it_ent, ger_ent, gre_ent, ser_ent)]
df['env'] = [np.nanmean([a,b,c,d]) for a, b, c, d in zip(it_env, ger_env, gre_env, ser_env)]

print(df)

fig = plt.figure()
ax=df.plot(kind='bar', x='year', rot=0, ylabel="MAE",xlabel="Year")
i = 0
for p in ax.patches:
    ax.annotate('%.2f'%(p.get_height()), (p.get_x() * 1.035, p.get_height() * 1.005),rotation=35)
    i += 1
ax.set_title("MAE per year for all cases")
```
